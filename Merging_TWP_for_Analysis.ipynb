```python
chems = pd.read_csv('Chemicals_for_filter.csv').chem_name.unique().tolist()
```

```python
def map_chem_casrn(chems, fd):
    chem_dic = {}
    for i, c in enumerate(chems):
        if c in fd.ChemicalName.unique().tolist():
            chem_dic[c] = fd.loc[fd.ChemicalName == c].CASRN.unique().tolist()
        elif c in fd.Chemical.unique().tolist():
            chem_dic[c] = fd.loc[fd.Chemical == c].CASRN.unique().tolist()
        else:
            chem_dic[c] = fd.loc[fd.Chemical == chems[i-1]].CASRN.unique().tolist()
    return chem_dic
```

```python
def get_ChemCASRN_df(chems, fd):
    
    chem_dic = map_chem_casrn(chems, fd)
    return pd.DataFrame.from_dict(data = chem_dic, orient='index', columns=['CASRN']).\
           reset_index().rename(columns={"index":'Chemical'})
```

```python
# Give the unique CASRN code to each chemical that was used for filtering.

chem_casrn = get_ChemCASRN_df(chems, fd1)
chem_casrn
```

```python
twp = pd.read_csv('twp_filter.csv')
twp
```

```python
# Merge the Newsreports with chemical-CASRN map, so that 
# we find the reports on the same chemical though having different expressions.

twp_chem_casrn = twp.merge(chem_casrn, on='Chemical')
twp_chem_casrn
```

```python
# Merge the ChemicalDates with Newsreports on unique code CASRN. 
# This is an all-in-one dataframe.

twp_chem_date = fd1.merge(twp_chem_casrn, on='CASRN')
twp_chem_date
```

```python
# Find the chemical that the time diff from publish date to scrape date 
# is less than a certain amount of years.

def find_target_reports(df, gap):
    '''
    Find the reports released no later than the gap years.
    gap: float, years after the scrape date.
    '''
    df['Year_Gap'] = (df['Date'] - df['Scrape Date'])/ pd.Timedelta(days=365)
    return df.loc[(df.Year_Gap >= 0) & (df.Year_Gap <= gap)]
```

```python
target = find_target_reports(twp_chem_date, 5)
target
```
